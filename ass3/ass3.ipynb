```python
import pandas as pd
import numpy as np

import re

Energy = pd.read_excel("assets/Energy Indicators.xls", usecols='C:F', skiprows=17, skipfooter=38)

columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']

Energy.columns = columns

for col in Energy.columns[1:]:
    Energy[col] = pd.to_numeric(Energy[col], errors='coerce')


Energy['Energy Supply'] = Energy['Energy Supply'].apply(lambda x: x * 10**6 if not pd.isna(x) else x)

Energy['Country'] = Energy['Country'].str.replace(r'\d+$| \(.*\)', '')

Energy['Country'].replace({"Republic of Korea": "South Korea", "United States of America": "United States", "United Kingdom of Great Britain and Northern Ireland": "United Kingdom", "China, Hong Kong Special Administrative Region": "Hong Kong"}, inplace=True)

GDP = pd.read_csv('assets/world_bank.csv', skiprows=4)

GDP.replace({"Korea, Rep.": "South Korea",  "Iran, Islamic Rep.": "Iran", "Hong Kong SAR, China": "Hong Kong"},inplace=True)

GDP.rename(columns={'Country Name' : 'Country'}, inplace=True)

ScimEn = pd.read_excel('assets/scimagojr-3.xlsx')

drop_cols = [str(i) for i in range(1960, 2006)]
for col in GDP.columns:
    if col in drop_cols:
        GDP.drop(col, inplace=True, axis=1)


gdp_col = ['Country','2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']

GDP = GDP[gdp_col]

mer1 = pd.merge(ScimEn, Energy, how='inner', on="Country")
res = pd.merge(mer1, GDP, how='inner', on="Country").set_index('Country')

res_15 = res[res['Rank'] <= 15]

res.head(n=20)
```

```python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'
                  }

res = res_15['PopEst'].groupby(by=ContinentDict).agg(['size', 'sum', 'mean', 'std'])


res.index.name = 'Continent'
res = res.reindex(['Asia', 'Australia', 'Europe', 'North America', 'South America'])

res.head()
```
